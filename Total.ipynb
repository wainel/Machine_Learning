```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
from tensorflow import feature_column
from sklearn.model_selection import train_test_split
```

```python
from google.colab import files
data_to_load = files.upload()
```

```python
import io
dataframe = pd.read_csv(io.BytesIO(data_to_load['trainingdata.csv']),sep=';')
dataframe.head()
```

```python
nb_rows, nb_cols = dataframe.shape
```

```python
df = pd.DataFrame(dataframe['TITLE_LENGTH'])
for col in dataframe.columns:
  if(dataframe[col].dtype == 'object'):
    serie = dataframe[col]
    one_hot = serie.unique()
    nb_one_hot = len(one_hot)
    if len(one_hot) < 10:
      one_hot_vector = np.zeros((nb_rows,nb_one_hot), dtype=int)
      for j in range(nb_one_hot):
        one_hot_vector[serie == one_hot[j],j] = 1
        new_title = col + '_' + one_hot[j]
        df[new_title] = one_hot_vector[:,j]
    else:
      serie = serie.str.replace(',','.')
      serie.astype(float)
      df[col] = serie
```
