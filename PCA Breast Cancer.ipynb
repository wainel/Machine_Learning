```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print("There are 2 different type of cancers: ", cancer['target_names'][0], " and ", cancer['target_names'][1])
```

```python
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
```

```python
df.head()
```

There are 30 features / columns. By using PCA, we'll try to reduce this 30 dimension into 2 dimension.

**Standard Scaling**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data=scaler.transform(df)
```

```python
print(scaled_data)
```

Tha dataset was converted into an array and now we have mean=1 and variance=0.

**Dimensionality reduction**

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
x_pca
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c=cancer['target'])
plt.xlabel('First principal component')
plt.ylabel('Second principal component')
```

In yellow, there are the malignant cancers and in purple, there are the benign cancers.
